```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_circles
from sklearn.metrics import accuracy_score
import plotly.graph_objects as go
```

```python
def initialization(X):
    W = np.random.randn(X.shape[1],1)
    b = np.random.randn(1)
    return (W, b)
```

```python
def model(X, W, b):
    Z = X.dot(W) + b
    A = 1 / (1 + np.exp(-Z))
    return A
```

```python
def log_loss(A, y):
    return 1/len(y) * np.sum(-y * np.log(A) - (1-y) * np.log(1-A))
```

```python
def gradients(A, X, y):
    dW = 1 / len(y) * np.dot(X.T, A-y)
    db = 1 / len(y) * np.sum(A-y)
    return (dW, db)
```

```python
def update(dW, db, W, b, learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return (W, b)
```

```python
def predict(X, W, b): 
    A = model(X, W, b)
    return A >= 0.5
```

```python
def artificial_neuron(X, y, learning_rate=0.1, n_iter=500):
    
    # Initilization W, b
    W, b = initialization(X)
    
    Loss = []

    #Train
    for i in range(n_iter):
        A = model(X, W, b)
        Loss.append(log_loss(A, y))
        dW, db = gradients(A, X, y)
        W, b = update(dW, db, W, b, learning_rate)
        
    y_pred = predict(X, W, b)
    print("accuracy score : ", accuracy_score(y, y_pred))
        
    plt.plot(Loss)
    plt.show()
    
    return (W, b)
        
    
    
```

```python
# Learning linear decision boundary of set of points

X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=0)
y = y.reshape((y.shape[0],1))

print("X dimensions : ", X.shape)
print("y dimensions : ", y.shape)

plt.scatter(X[:,0], X[:,1], c=y, cmap='cool')
plt.show()
```

```python
W,b = artificial_neuron(X, y)
```

```python
new_plant = np.array([2,1])
x0 = np.linspace(-1,4,100)
x1 = (-W[0]*x0 - b) / W[1]

plt.scatter(X[:,0], X[:,1], c=y, cmap='cool')
plt.scatter(new_plant[0], new_plant[1], c='black')
plt.plot(x0, x1, c='orange', lw=3)
plt.show()
print("Label of new data point is : ", predict(new_plant, W, b))
```
